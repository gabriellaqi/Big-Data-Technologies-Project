```python
# Initiate Spark Context
from pyspark import SparkContext, SparkConf
import re
conf = SparkConf().setAppName("appName")
sc = SparkContext(conf = conf)
```

```python
doc = sc.textFile("en_US.blogs.txt")
```

```python
flattened = doc.filter(lambda line : len(line) > 0) \
.flatMap(lambda line:re.split('W+', line))
kvPairs = flattened.filter(lambda word : len(word) > 0) \
.map(lambda word: (word.lower(), 1))
```

```python
countsByWord = kvPairs.reduceByKey(lambda v1, v2 : v1 + v2) \
.sortByKey(ascending = False)
topWords = countsByWord.map(lambda x: (x[1], x[0])) \
.sortByKey(ascending = False)
```

```python
words = doc.flatMap(lambda x:x.split(' ')) \
.filter(lambda x : len(x) > 0)
sorted = words.sortBy(lambda x: x[0].lower(), ascending = False)
dist = sorted.distinct()
```

```python
WordsCount=doc.flatMap(lambda line: line.split(" ")).map(lambda word: (word, 1))
WordsCount.count()
DistinctWordsCount=WordsCount.reduceByKey(lambda a,b: a+b)
DistinctWordsCount.count()
SortedWordsCount=DistinctWordsCount.map(lambda a: (a[1], a[0])).sortByKey()
SortedWordsCount.top(100)
```

```python
Sorted_final=SortedWordsCount.map(lambda a: (a[1], a[0]))
Sorted_final.top(40)
```
